# Basic Image Manipulation

```python
import cv2 as cv
import matplotlib.pyplot as plt
from IPython.display import Image
%matplotlib inline
```

```python
fox_image = cv.imread("../images/fox-8x7.png", cv.IMREAD_COLOR_RGB)
fox_image_gray = cv.cvtColor(fox_image, cv.COLOR_RGB2GRAY)

print(fox_image_gray.shape)
print(fox_image_gray)

plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.title("Fox Image")
plt.imshow(fox_image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Fox Image Grayscale")
plt.imshow(fox_image_gray, cmap='gray')
```

## Accessing Individual Pixels

```python
print(fox_image_gray[4, 0])
print(fox_image_gray[4, 1]) 
print(fox_image_gray[4, 2])
```

## Modifying Image Pixels

```python
fox_image_gray[3, 3] = 0
fox_image_gray[3, 4] = 0

plt.imshow(fox_image_gray, cmap='gray')
```

## Cropping Images

```python
landscape = cv.imread("../images/landscape.jpg", cv.IMREAD_COLOR_RGB)
plt.imshow(landscape)
```

### Crop out the Morro do Careca

```python
landscape_crop = landscape[300:400, 0:350]
plt.imshow(landscape_crop)
```

## Resizing Images

We can resize images with 2 methods, using scaling factor or specifyng exact size

### Method 1: Specifying Scaling Factor using fx and fy

```python
resized_landscape_crop_2x = cv.resize(landscape_crop, dsize=None, fx=2, fy=2)
plt.imshow(resized_landscape_crop_2x)
```

### Medthod 2: Specifying exact size of the output image

```python
desired_width = 1280
desired_height = 720
dim = (desired_width, desired_height)

resized_landscape_crop = cv.resize(landscape_crop, dsize=dim)
plt.imshow(resized_landscape_crop)
```

## Flipping Images

```python
landscape_flip_horizontal = cv.flip(landscape, 1)
landscape_flip_vertical = cv.flip(landscape, 0)
landscape_flip_both = cv.flip(landscape, -1)

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.title("Original Landscape")
plt.imshow(landscape)

plt.subplot(2, 2, 2)
plt.title("Landscape Flip Horizontal")
plt.imshow(landscape_flip_horizontal)

plt.subplot(2, 2, 3)
plt.title("Landscape Flip Vertical")
plt.imshow(landscape_flip_vertical)

plt.subplot(2, 2, 4)
plt.title("Landscape Flip Both")
plt.imshow(landscape_flip_both)
```
