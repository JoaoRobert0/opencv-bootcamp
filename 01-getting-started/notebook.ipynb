# Getting started

```python
import cv2 as cv
import matplotlib.pyplot as plt
from IPython.display import Image
%matplotlib inline
```

## How to read images using OpenCV

```python
image = cv.imread('../images/checkboard-84x84.jpeg', flags=cv.IMREAD_GRAYSCALE)
print(image)
```

**flags=""**
- IMREAD_GRAYSCALE or 0: To read in grayscale
- IMREAD_COLOR or 1: To read in BGR

## Display images with the real size

```python
Image(filename='../images/checkboard-84x84.jpeg')
```

## Display images using Matplolib

```python
plt.imshow(image)
```

This error happened because the OpenCV read a image in BGR but the Matplotlib expect a image in RGB format. But in this case we can pass the arg 'gray' to the parameter cmap=''

```python
plt.imshow(image, cmap='gray')
```

## Convert to different color spaces

```python
image_bgr = cv.imread('../images/mcdonald.png', flags=cv.IMREAD_COLOR)
plt.imshow(image_bgr)
```

```python
image_rgb = cv.cvtColor(image_bgr, cv.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

## Splitting and Merging color channels

```python
landscape = cv.imread('../images/landscape.jpg', flags=cv.IMREAD_COLOR)
b, g, r = cv.split(landscape)

plt.figure(figsize=(15, 5))

plt.subplot(141)
plt.title('Blue Channel')
plt.imshow(b, cmap='gray')

plt.subplot(142)
plt.title('Green Channel')
plt.imshow(g, cmap='gray')

plt.subplot(143)
plt.title('Red Channel')
plt.imshow(r, cmap='gray')

merged_image = cv.merge((r, g, b))
plt.subplot(144)
plt.title('Merged Image')
plt.imshow(merged_image)
```

## Saving images in disk

```python
cv.imwrite('landscape-merged.jpg', cv.cvtColor(merged_image, cv.COLOR_RGB2BGR))
Image(filename='landscape-merged.jpg')
```
